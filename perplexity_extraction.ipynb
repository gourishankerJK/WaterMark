```python
import pandas as pd
import matplotlib.pyplot as plt


def read_csv_file(file_path):
    return pd.read_csv(file_path, header=0, encoding='utf-8')

old_data = read_csv_file('Per_ult1.csv')
new_data = read_csv_file('Per_ult1_ultimate.csv')
```

```python
old_data = old_data[[i for i in old_data.columns if 'per' in i]]
new_data = new_data[[i for i in new_data.columns if 'per' in i]]
```

```python
old_data
```

```python
data = pd.concat([old_data, new_data], axis=1)
data = data.reset_index(drop=True)
data
```

```python
changed = ['Watermarked Text', 'RP0', 'RP1', 'RP2', 'RP3', 'RP4', 'PT', 'RT', 'RP0 + RT', 'RP1 + RT', 'RP2 + RT', 'RP3 + RT', 'RP4 + RT', 'RT + Para', 'PT + Para']
a = ['per_of_en_kwg_0', 'per_of_RecursivePara_kwg_0', 'per_of_RecursivePara_kwg_1', 'per_of_RecursivePara_kwg_2', 'per_of_RecursivePara_kwg_3', 'per_of_RecursivePara_kwg_4', 'per_of_PivotTranslation_kwg_0', 'per_of_NormalTranslation_kwg_0', 'per_of_RERERETE_kwg_0', 'per_of_RERERETE_kwg_1', 'per_of_RERERETE_kwg_2', 'per_of_RERERETE_kwg_3', 'per_of_RERERETE_kwg_4', 'per_of_RNormalTranslation_kwg_0', 'per_of_RPivotTranslation_kwg_0']
print(a)
df = data[a]
plt.figure(figsize=(16, 6))
plt.tick_params(axis='x', width=0.5)
boxplot = plt.boxplot(df[a], labels=changed,  patch_artist=True, widths=0.45)
plt.ylim(0, 200)  
plt.ylabel("Perplexity")
plt.xlabel("Watermarking Techniques")
plt.savefig('Dataset/Graphs/perp.png')
plt.show()
```
